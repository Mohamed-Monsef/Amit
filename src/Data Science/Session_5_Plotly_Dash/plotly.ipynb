```python
import plotly.express as px
import plotly.subplots as sp
import plotly.graph_objects as go
 
# Create subplot layout: 2 rows Ã— 3 columns
fig = sp.make_subplots(rows=2, cols=3, subplot_titles=cat_cols,
                       specs=[[{'type':'domain'}]*3, [{'type':'domain'}]*3])
 
for i, col in enumerate(cat_cols):
    row = i // 3 + 1
    col_pos = i % 3 + 1
   
    unique = df[col].value_counts()
    categories = unique.index
    count = unique.values
   
    fig.add_trace(
        go.Pie(labels=categories, values=count, name=col,
               hole=0,  # change to 0.3 for donut style
               textinfo="percent+label"),
        row=row, col=col_pos
    )
 
# Layout adjustments
fig.update_layout(
    height=600, width=1000,
    title_text="Categorical Columns - Pie Charts",
    legend=dict(x=1, y=0.5)
)
 
fig.show()
```

```python
import pandas as pd
import seaborn as sns
import plotly.express as px
 
# Load Titanic dataset
titanic = sns.load_dataset('titanic')
 
# Keep numeric columns and drop NaNs
data = titanic[['age', 'fare', 'sibsp', 'pclass']].dropna()
 
# Detect outliers using IQR
def detect_outliers(df, cols):
    outlier_mask = pd.Series(False, index=df.index)
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        outlier_mask |= (df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)
    return outlier_mask
 
data['outlier'] = detect_outliers(data, ['age', 'fare', 'sibsp'])
 
# Map outlier flag to string for better coloring
data['type'] = data['outlier'].map({True:'Outlier', False:'Normal'})
 
# 3D animated scatter
fig = px.scatter_3d(
    data,
    x='age', y='fare', z='sibsp',
    color='type',       # Normal vs Outlier
    symbol='type',
    animation_frame='pclass',  # Animate by passenger class
    size_max=10,
    title='Titanic 3D Scatter with Outliers Animated by Passenger Class'
)
 
fig.show()
 
 
 
 
```
