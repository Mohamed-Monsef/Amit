###  To fix this issue You will fill these values with ........... for numeric and ......for categorical

```python
#Numerical Data.

birth_year_median = df['member_birth_year'].median()
df['member_birth_year'] = df['member_birth_year'].fillna(birth_year_median)


#Categorical Data.
null_cols = ["start_station_id",
             "end_station_id",
             "start_station_name",
             "end_station_name",
             "member_gender"]

for col in null_cols:
    mode_value = df[col].mode()[0]
    df[col] = df[col].fillna(mode_value)
    
    
# Converting Float numbers into Int because it looks messy. 
df['member_birth_year'] = df['member_birth_year'].astype(int)
df['start_station_id'] = df['start_station_id'].astype(int)
df['end_station_id'] = df['end_station_id'].astype(int)

```

```python
#check the Nulls

print(df.isnull().sum())
```

```python
##  step  2 ) Incorrect data type as in (( `user_type` &&	`member_gender`)) :  significant change , so  need to clean it.
df["user_type"] = df["user_type"].astype("category")
df["member_gender"] = df["member_gender"].astype("category")
```

```python
## check your work 
df.info()
```

```python
## step 3 ) Good factor for marketing work as in ((`member_birth_year`)) : so need to clean it.
## To know the target audience's age and serve on them
# create a column for users' age depending on the current year

## since the dataset was in Feb 2019 we should make the data depend on this date

df["age"] =  2019 - df["member_birth_year"]
df[["age","member_birth_year"]].head()
```

```python
# check your work 
df.info()
```

```python
## change the new column age data type from float to int

df["age"] = df["age"].astype(int)
```

```python
df.info()
```

```python
## create a function to know the average users ages
def average_users_age():
    
    """
    This function calculates  the average users ages
    """
    users_age_mean =  df["age"].mean()
    return users_age_mean
avg = average_users_age()
print(f"The average users ages is: {avg}")
```

```python
## another important factor to offer  marketing services is the trip duration time 

## need to change the time from second to minute to be more usable

df["duration_min"] = (df["duration_sec"] / 60).round(2)
df[["duration_min","duration_sec"]].head()
```

```python
## I need to change the time from second to Houre  to be more usable
df["duration_hour"] = (df["duration_sec"] / 3600).round(2)

# Check the results
print(df[["duration_sec", "duration_hour"]].head())
```

# <a id="13">  13 ) . What is the  Correlation between each variable in the dataset?</a>

```python
# Make visualization to easy get insight
## check your work 
plt.figure(figsize= (10,5), dpi = 150)
sns.heatmap(df.corr(numeric_only=True), annot = True, fmt = '.2f', cmap = 'vlag_r', center = 0)
plt.title("Represents Correlation between each variable  in the dataset")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```


# <a id="14"> 14 )  .  What is the  Correlation between `specific numeric variables` in the dataset?</a>


```python
# Make visualization to easy get insight

numeric_vars = ['age', 'duration_min', 'start_station_latitude', 'start_station_longitude']

# Calculate the correlation matrix
correlation = df[numeric_vars].corr()

# Plot the Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Specific Numeric Variables')
plt.show()
```

```python
data_stat = ['age', 'duration_min',"duration_sec", "member_birth_year", "bike_id"]
plot = sns.PairGrid(data = df, vars = data_stat ) 
plot.map(plt.scatter, alpha = 1/4)
```
