```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.array([1,2,3,4])
y= np.array([2,2.8,3.6,4.5])
```

```python
w = 0
b = 0
alpha = 0.01
num_iteration = 20
```

```python
sse_values = []

#Gradient descent

for i in range(num_iteration):
    
    y_hat = w*x + b 
    ####gradient of w 
    D_w = 2 * np.sum((y_hat -y)*x)
    ### gradient of  b 
    D_b = 2 * np.sum(y_hat-y)
    
    #Update parameters
    w -= alpha * D_w
    b -= alpha * D_b
    
    #compute sse and store it on sse variable
    sse = (np.sum((y_hat -y)**2))
    sse_values.append(sse)
    
    if (i+1)  % 20 == 0 :
        print(f"iteration {i+1}, SSE: {sse}")
    
    
```

```python

print(f"Optimized parameters: bais = {b}, slop = {w}")
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(range(num_iteration),sse_values,label='sse')
plt.xlabel("Iteration")
plt.ylabel('SSE')
plt.title('SEE over iteration')
plt.legend()
# plt.show()

plt.subplot(1,2,2)
plt.scatter(x,y,color="blue",label='Data Points')
plt.plot(x, b+w*x, color = 'red', label='Regression Line')
plt.xlabel("X")
plt.ylabel('Y')
plt.title("Linear Regression Fit line")
plt.legend()
plt.show()
```
