```python
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
#Data in (datadata) txt file
X = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])

# linear Regression 
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_lin = lin_reg.predict(X)

# Ridge Regression with small alpha
ridge = Ridge(alpha=1.0)  
ridge.fit(X, y)
y_pred_ridge = ridge.predict(X)

# Lasso Regression with larger alpha
lasso = Lasso(alpha=10.0)   
lasso.fit(X, y)
y_pred_lasso = lasso.predict(X)


print("Linear Regression:")
print(" MSE:", mean_squared_error(y, y_pred_lin))
print(" R²:", r2_score(y, y_pred_lin))
print()

print("Ridge Regression:")
print(" MSE:", mean_squared_error(y, y_pred_ridge))
print(" R²:", r2_score(y, y_pred_ridge))
print()

print("Lasso Regression:")
print(" MSE:", mean_squared_error(y, y_pred_lasso))
print(" R²:", r2_score(y, y_pred_lasso))
```

```python
# Linear Regression
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_pred_lin, color="red", label="Linear Regression")
plt.title("Linear Regression")
plt.legend()
```

```python
#Ridge Regression
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 2)
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_pred_ridge, color="green", label="Ridge Regression")
plt.title("Ridge Regression")
plt.legend()
```

```python
# Lasso Regression
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 3)
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_pred_lasso, color="purple", label="Lasso Regression")
plt.title("Lasso Regression")
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# Comparison Scatter Plot 
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_pred_lin, color="red", label="Linear Regression")
plt.plot(X, y_pred_ridge, color="green", linestyle="--", label="Ridge Regression")
plt.plot(X, y_pred_lasso, color="purple", linestyle=":", label="Lasso Regression")
plt.title("Comparison of Linear, Ridge, and Lasso Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
