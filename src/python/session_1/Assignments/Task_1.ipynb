```python
#2.1 Encoded Message: ###!!@mocleW EPGTQ!!!6789
encoded_message_1= "###!!@mocleW EPGTQ!!!6789"
Extract_message_1= encoded_message_1.split("@")[1].split("!")[0]
print(Extract_message_1)
```

```python
#2.2 Reverse the first word: "mocleW" becomes "Welcome".
reversedd_0 = Extract_message_1.split()
reversedd_1 = reversedd_0[0]
reversedd = "".join(reversed(reversedd_1))
print(reversedd)

'''
Wrong word from "mocleW" to "Welcome"  missing E in melcoW  
'''
```

```python
#2.3 Replacing
### the sentence contains a vowel
replace_0 =reversedd_0[1].replace("EPGTQ","PGTQ")
print(replace_0)
```

```python
#2.4 final decoded 
print(f"{reversedd} {replace_0}")
```
