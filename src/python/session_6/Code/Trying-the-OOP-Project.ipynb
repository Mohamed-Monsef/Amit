```python
class Student:
    
    _id_counter = 1
    
    def __init__(self,name):
        self.student_id = Student._id_counter
        Student._id_counter += 1
        self.name = name
        self.grades = {}
        self.enrolled_courses = []
        
    def add_grade(self,course_id,grade):
        """
        Adds or updates a grade for a specific course ID.
        """
        if not 0 <= grade <= 100:
            raise ValueError("Please enter a value between 0 and 100!")
        self.grades [course_id] = grade
        
    def enroll_in_course(self,course):
        """
        Adds course to student enrollment list.
        """
        if course not in self.enrolled_courses:
            
            self.enrolled_courses.append(course)
        else:
            print("Your course is already enrolled")
            
    def __str__(self):
        return f"Student ID: {self.student_id},Name: {self.name} and Grade: {self.grades}"
    def __repr__(self):
        return self.__str__()
        
```

```python
# Create objects
s1 = Student("Alice")
s2 = Student("Bob")

print(s1)  # Should show ID: 1
print(s2)  # Should show ID: 2
```

```python
s1.add_grade("ML",100)
```

```python
s1.enroll_in_course("Python 101")
s1.enroll_in_course("SQL")
print(f"Enrolled Courses: {s1.enrolled_courses}")
```
