```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def move(self):
        pass

class Bird():
    def move(Animal):
        print("Move from Bird")
    

class Cat():
    def move(Animal):
        print("Move from Cat")
        
```

```python
a = Bird()
a.move()
```

```python
class Calculator1:
    
    def sum(self,n1,n2):
        return n1+n2
    
    def sub(self,n1,n2):
        return n1-n2
    
    def multi(self,n1,n2):
        return n1 * n2
    
    def div(self,n1,n2):
        return n1 / n2
    
class Calculator2(Calculator1):
    
    def power(self,n1,n2):
        return n1 ** n2
    
c1 = Calculator2()

print(c1.sum(5,6))

        
    
        
    
```

```python
from prettytable import PrettyTable

products = [{"name":"water","price":80.0,"quantity":1500},
            {"name":"soda","price":130.0,"quantity":1500},
            {"name":"chips","price":75.0,"quantity":1500},
            {"name":"bread","price":45.0,"quantity":1500},
            {"name":"eggs","price":200.0,"quantity":1500}
            ]

table = PrettyTable()
table.field_names = ["Product Name", "Price($)","quantity"]
for p in products:
    table.add_row([p["name"],f"{p["price"]:.2f}",p["quantity"]])
print(table)

total_bill = 0

while True:
    product_name = input("\n Enter Product name or exit to finish:").strip().lower()
    if product_name ==  "exit":
        break
    product = next((p for p in products if p["name"].lower() == product_name), None)
    if product is None:
        print("Product Not Found, Please Try Again.")
        continue
    try:
        quantity_required = float(input(f"How many  {product["name"]} do you want?"))
        if quantity_required <= 0:
            print("Quantity must be greater than 0")
            continue
        if quantity_required > product["quantity"]:
            print(f"Insufficient Quantity!")
            continue
        discount_percentage = 0.05
        price_before_discount = quantity_required * product["price"]
        discount_amount = price_before_discount * discount_percentage
        final_price = price_before_discount - discount_amount
        
        product["quantity"] -= quantity_required
        total_bill += final_price
        print(f"Success! price for{product["name"]}: ${final_price:.2f}(saved: ${discount_amount:.2f})")
        another = input("something else? (y/n)").lower()
        if another !="y":
            break
    except ValueError:
        print("Invalid input,Please enter number of Quantity!")
print(f"\n Total Amount to pay: ${total_bill}")
print("Thank you, Have a Great day.")
```
