```python

def roman_to_int(s: str) -> int:
        # Map for lookup
    roman_map = {
        "I": 1, "V": 5, "X": 10, "L": 50,
        "C": 100, "D": 500, "M": 1000
    }
        
    total = 0
    n = len(s)
        
    for i in range(n):
        if i < n - 1 and roman_map[s[i]] < roman_map[s[i+1]]: ######  لو الرقم اكبر من الرقم اللي جنبيه اجمع واحد علي سبيل المثال VI
                ### لو الرقم اقل من الرقم ا للي حنبيه اطرح علي سبيل المثال IV
                
            total -= roman_map[s[i]]
        else:
            total += roman_map[s[i]]
                
    return total
    
user_input = input("Please Enter Your Roman Words: ").upper()
print(f"Result: {roman_to_int(user_input)}")
```
