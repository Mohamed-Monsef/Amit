```python
import os
import random

def thanos_func(folder):
    # 1. Ensure the folder exists
    if not os.path.exists(folder):
        os.makedirs(folder)
        print(f"Created folder: {folder}")

    # 2. Create a number of files inside it(20 file)
    for i in range(1, 21):
        file_name = f"universe_citizen_{i}.txt"
        with open(os.path.join(folder, file_name), 'w') as f:
            f.write("Mohamed Mostafa Abdelmonsef")
    
    # 3. Check the len of files in that folder
    files = os.listdir(folder)
    initial_count = len(files)
    print(f"Total files before the func: {initial_count}")

    # 4. Delete random half of this folder
    # We calculate half the length of the list
    half_count = initial_count // 2
    
    # Randomly select deleted_files
    deleted_files = random.sample(files, half_count)
    
    for deleted in deleted_files:
        file_path = os.path.join(folder, deleted)
        os.remove(file_path)
        print(f"Deleted: {deleted}")

    # 5. Check the result
    remaining_files = os.listdir(folder)
    print("--- func Complete ---")
    print(f"Files remaining: {len(remaining_files)}")
    print(f"Survivors: {remaining_files}")

# Run the function
thanos_func("the_universe")
```
