# Problem Description – Task 1 (Python Calculator Validation)

## Task Description

- Modify the provided calculator code to validate user inputs using **try-except** blocks.  
- Handle the following cases:  
  1️⃣ Non-numeric input → print a clear error message.  
  2️⃣ Division by zero → print a clear error message.  
  3️⃣ Any unexpected errors → print a general message with the error description.

## Notes
- Make sure function names are correct (e.g., `multiply` instead of `multply`).  
- The goal is to make the program safer and user-friendly, without crashing on errors.


```python
def sum(num_1, num_2):
    
    return num_1 + num_2

def sub(num_1, num_2):
    
    return num_1 - num_2
def div(num_1, num_2):
    
    return num_1 / num_2
def multiply(num_1, num_2):
    
    return num_1 * num_2

def main():
    print("---------calc------")
    print("choose operations")
    print("1-sum")
    print("2-sub")
    print("3-div")
    print("4-mul")
    
    choice = input("Enter the operation")
    if choice not in [ "1","2","3","4"]:
        print("Invalid input, Please choose between the first four numbers")
        return
    try:
        num_1 = float(input("Enter first number"))
        num_2 = float(input("Enter second number"))
        
        if choice == "1":
            result = sum(num_1, num_2)
        elif choice == "2":
            result = sub(num_1, num_2)
            
        elif choice == "3":
            result = div(num_1, num_2)
            
        elif choice == "4":
            result = multiply(num_1, num_2)
            
    except ValueError:
        print("Error, Please add a Numeric Value")
    except ZeroDivisionError:
        print("Error, you can not divide by zero")
    
    print(f"RESULT: , {result}")
if __name__ == "__main__":
    main()
```
